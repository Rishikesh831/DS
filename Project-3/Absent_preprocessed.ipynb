## Data loading

## Data exploration

### Subtask:
Explore the loaded dataset to understand its structure, identify potential issues like missing values, and examine data types.  Pay close attention to the 'Reason for Absence' column.


**Reasoning**:
Explore the dataset by examining its shape, data types, missing values, and descriptive statistics, focusing on the 'Reason for Absence' column and the 'Date' column.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Re-examine data types and check for null values; investigate 'Reason for Absence'
print("\nDataFrame Information:")
display(df.info())
print("\nValue counts for 'Reason for Absence':")
display(df['Reason for Absence'].value_counts())

# Descriptive statistics for numerical features
print("\nDescriptive Statistics:")
display(df.describe())

# Investigate the 'Date' column
print("\nInformation about the 'Date' column:")
print("Data type of 'Date' column:", df['Date'].dtype)
# Convert 'Date' column to datetime objects
try:
    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
    print("Successfully converted 'Date' column to datetime objects.")
except ValueError as e:
    print(f"Error converting 'Date' column to datetime: {e}")

# Explore the distribution of 'Absenteeism Time in Hours'
print("\nDistribution of 'Absenteeism Time in Hours':")
display(df['Absenteeism Time in Hours'].describe())
```

## Data cleaning

### Subtask:
Check for and handle any inconsistencies or duplicates in the data.  Specifically, investigate the `Reason for Absence` column further, and address any issues related to it.  Also ensure that the data types of all columns are correct and consistent with the expected data.


**Reasoning**:
Investigate the 'Reason for Absence' column for inconsistencies, check for duplicates, verify data types, and optionally handle outliers.



```python
# Investigate 'Reason for Absence'
print("Unique values in 'Reason for Absence':", df['Reason for Absence'].unique())
print("Number of unique values:", df['Reason for Absence'].nunique())

# Check for duplicates
duplicates = df.duplicated()
num_duplicates = duplicates.sum()
print(f"\nNumber of duplicate rows: {num_duplicates}")
if num_duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicate rows removed.")

# Verify Data Types
print("\nData Types before correction:")
display(df.dtypes)

# Check if 'Reason for Absence' needs to be categorical
if pd.api.types.is_numeric_dtype(df['Reason for Absence']):
    df['Reason for Absence'] = df['Reason for Absence'].astype('object')
    print("'Reason for Absence' converted to object type.")

print("\nData Types after correction:")
display(df.dtypes)


# Address Outliers (Optional):
# For this example, we will cap 'Absenteeism Time in Hours' at 3 standard deviations above the mean
# This is just one approach. Other methods could include winsorization or removing outliers.
upper_limit = df['Absenteeism Time in Hours'].mean() + 3 * df['Absenteeism Time in Hours'].std()
df['Absenteeism Time in Hours'] = df['Absenteeism Time in Hours'].clip(upper=upper_limit)
print(f"\n'Absenteeism Time in Hours' capped at {upper_limit:.2f}")
```

## Data wrangling

### Subtask:
Transform the 'Reason for Absence' column into dummy variables, and perform other necessary data transformations.


**Reasoning**:
Transform the 'Reason for Absence' column into dummy variables, group them, drop the original column, and perform other necessary data transformations as instructed.



```python
# Create dummy variables for 'Reason for Absence'
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)

# Group the dummy variables into four categories
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)

# Concatenate the grouped dummy variables with the original DataFrame
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)

# Rename the new columns
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
                'Daily Work Load Average', 'Body Mass Index', 'Education',
                'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns = ['ID', 'Reason for Absence'] + column_names

# Drop the original 'Reason for Absence' column
df = df.drop(['Reason for Absence'], axis=1)
df = df.drop(['ID'], axis=1)


# Map education levels to numerical values
df['Education'] = df['Education'].map({1:0, 2:1, 3:1, 4:1})

# Extract month from the 'Date' column
df['Month Value'] = df['Date'].dt.month

# Reorder columns (optional, but good practice)
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value'] + [col for col in df.columns if col not in ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value']]
df = df[column_names_reordered]


# Display the preprocessed DataFrame
display(df.head())
```
