```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Telco.csv')
```

```python
df.info()
df.describe()
df.columns
```

```python
df =df.drop('customerID',axis=1)
```

```python
df.head()
```

```python
df['TotalCharges'] = df['TotalCharges'].str.strip()
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.dropna(inplace=True)
df['TotalCharges'].value_counts()
```

```python
for col in df.select_dtypes(include='object'):
    print(f"{col}: {df[col].unique()}")
cols_to_fix = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
               'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in cols_to_fix:
    df[col] = df[col].replace({'No internet service': 'No'})
```

```python
for col in df.select_dtypes(include='object'):
    print(f"{col}: {df[col].unique()}")
cols_to_fix = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
               'TechSupport', 'StreamingTV', 'StreamingMovies']
```

```python
binary_map = {'Yes':1,'Male':1,'Female':0,'No':0}
df['gender']=df['gender'].map(binary_map)
df['Partner']=df['Partner'].map(binary_map)
df['Dependents']=df['Dependents'].map(binary_map)
df['PhoneService']=df['PhoneService'].map(binary_map)
df['Churn'] = df['Churn'].map(binary_map)
df['PaperlessBilling']=df['PaperlessBilling'].map(binary_map)
cols_to_map = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
               'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in cols_to_map:
    df[col] = df[col].map(binary_map)
```

```python
df.head()
```

**MULTIPLE LINES HAS 3 CATEGORIES WHICH ARE NUMERICAL RELATED SO WE SET THEM AS 0,1 AND 2**

```python
# Map according to meaning
multiple_lines_map = {
    'No phone service': 0,  # 0 lines
    'No': 1,                # 1 line
    'Yes': 2                # Multiple lines
}

df['MultipleLines'] = df['MultipleLines'].map(multiple_lines_map)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['InternetService', 'Contract', 'PaymentMethod'], drop_first=True)
```

```python
df.head()
```

```python
df = df.astype({col: 'int' for col in df.select_dtypes('bool').columns})
```

```python
df.head()
```
