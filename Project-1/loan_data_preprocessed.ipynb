```python
df = pd.read_csv('train_loan.csv')
df.head()
```

```python
df['EMI'] = (df['LoanAmount']*1000/df['Loan_Amount_Term']).round(2)
df.head()
```

You may have noticed the actual EMI formula is not used that is because it is unneccesary to factor in interest rate while we just want to filter out applicants based on this column

```python
df.drop(columns=['LoanAmount','Loan_Amount_Term','Loan_ID','Gender','Married'],axis=1,inplace=True)
df.head()
```

```python
df['Dependents'] = df['Dependents'].replace('3+', 3).astype(int)
df['Self_Employed']= df['Self_Employed'].replace({'Yes':1,'No':0})
df['Education']= df['Education'].replace({'Graduate':1,'Not Graduate':0})
df['Loan_Status']= df['Loan_Status'].replace({'Y':1,'N':0})
df = pd.get_dummies(df, columns=['Property_Area'], drop_first=True)
df.head()
```

```python
df[['Property_Area_Urban', 'Property_Area_Semiurban']] = df[['Property_Area_Urban', 'Property_Area_Semiurban']].astype(int)
df.head()
```

```python
df.to_csv("loan_data_processed.csv", index=False)
```
